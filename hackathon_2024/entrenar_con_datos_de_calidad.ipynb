## Intro
### Beneficios de entrenar en corpus/datasets de calidad

- **Alineación y Adquisición de Conocimientos**: Los datos de alta calidad aseguran que el modelo aprenda información precisa y comportamientos apropiados. Esto es crucial para enseñar al modelo a alinearse con objetivos humanos y adquirir conocimientos durante el preentrenamiento.

- **Rendimiento y Generalización**: Entrenar con datos de baja calidad puede afectar negativamente el rendimiento de los LLMs. Los datos de alta calidad ayudan a reducir sesgos y mejorar la fiabilidad y capacidad de generalización del modelo.

- **Eficiencia y Escalabilidad**: La calidad de los datos afecta significativamente la eficiencia del proceso de entrenamiento. Datos de alta calidad permiten una escalabilidad más eficiente de los parámetros del modelo y pueden conducir a un modelo más eficiente en términos de cómputo.

- **Diversidad de Datos y Experticia en el Dominio**: Un conjunto de datos diverso y de alta calidad permite a los LLMs adquirir un amplio rango de conocimientos y demostrar fuertes habilidades de generalización en varios dominios. Los conjuntos de datos especializados también pueden proporcionar a los LLMs habilidades específicas para resolver problemas.

- **Preprocesamiento y Gestión de Datos**: La gestión efectiva de datos, incluyendo estrategias de preprocesamiento como la deduplicación, filtrado de calidad y filtrado de toxicidad, es esencial para construir conjuntos de entrenamiento adecuados. Este proceso influye significativamente en la capacidad del modelo para aprender de manera eficiente y desempeñarse efectivamente en diferentes tareas.

- **Direcciones Futuras y Desafíos**: Hay un enfoque creciente en mejores conjuntos de datos para el preentrenamiento en lugar de entrenar con todos los datos disponibles. El equilibrio entre la calidad y la cantidad de datos sigue siendo un área de investigación activa. Los datos de alta calidad pueden permitir entrenar modelos más pequeños que se desempeñan de manera comparable a modelos más grandes, pero existen riesgos como el sobreajuste y la sensibilidad al prompting.

## Pre-training

### Procedimiento de preparación de datos

![image](https://huggingface.co/mrm8488/images/resolve/main/Captura%20de%20pantalla%202024-03-11%20a%20las%2013.30.37.png?download=true)


- **Visión General del Procesamiento de Datos**
  - El objetivo es crear datos de entrenamiento bilingües (inglés y chino) de alta calidad a partir de la web.
  - Esto implica usar herramientas específicas para identificar idiomas y limpiar los datos.

- **Pasos de Limpieza**
  - **Filtros Heurísticos:** Eliminar texto de mala calidad basado en:
    - URLs malas, dominios y texto sin sentido.
    - Longitud de documento inapropiada y demasiados símbolos especiales o líneas incompletas.
    - Contenido repetitivo.
    - Oculta información personal como correos electrónicos y números de teléfono.
  - **Filtros Aprendidos:** Usar métodos avanzados para tratar contenido complicado, especialmente para chino:
    - Descartar documentos que son demasiado confusos o de baja calidad.
    - Eliminar documentos que son incoherentes o contienen contenido peligroso.
  - **Filtros Basados en Agrupaciones:** Agrupar documentos similares para evaluar y mejorar la calidad de manera más efectiva.
  - **Deduplicación:** Eliminar contenido duplicado para asegurar que cada documento sea único.

- **Resultado**
  - Produce un conjunto de datos con 3.1 billones de palabras de alta calidad en inglés y chino.
  - Este conjunto de datos es más limpio y diverso en comparación con otros, en parte debido a sus métodos efectivos de limpieza y filtrado.
